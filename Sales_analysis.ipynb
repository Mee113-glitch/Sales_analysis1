#### what product are most often sold together

```python
all_data.head(10)

#IF ORDER ID ARE SAME, THAT MEANS PRODUCT ARE SOLD TOGETHER
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df.head(10)
```

```python
# putting all products having same ID in 1 line
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head()
```

```python
#drop duplicates
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(10)
```

```python
#new lib
from itertools import combinations
from collections import Counter
count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3))) # you can keep it as 2 or 3 or anything
    #it just gives no of product bought together, 2 would give only 2

print(count)
```

```python
count.most_common(10)
```

```python
for key, value in count.most_common(10):
    print(key, value)
```

```python
# Assuming 'Quantity Ordered' is in a string format, you need to convert it to numeric.
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'], errors='coerce')
pro_group = all_data.groupby('Product')
quan_ordered = pro_group['Quantity Ordered'].sum()
products = [product for product, df in pro_group]
plt.bar(products, quan_ordered)
plt.xticks(products, rotation = 'vertical', size=8)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.show()
```

```python
prices = all_data.groupby('Product')['Price Each'].mean()
print(prices)
```

```python
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quan_ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
